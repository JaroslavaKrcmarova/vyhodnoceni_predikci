> # üìä Z√°kladn√≠ metody vyhodnocen√≠ predikc√≠ prodeje
Vyhodnocen√≠ predikce prodeje je **porovn√°n√≠ odhadovan√Ωch hodnot s re√°ln√Ωmi** prodeji, jeho c√≠lem je zjistit:

>- jak p≈ôesn√° byla p≈ôedpovƒõƒè,
>- kde vznikly chyby,
>- jak model do budoucna zlep≈°it.

## 1. Porovn√°n√≠ predikce vs. realita
| Obdob√≠ | Predikce | Skuteƒçnost | Odchylka |
| ------ | -------- | ---------- | -------- |
| leden  | 100 ks   | 90 ks      | -10 ks   |
| √∫nor   | 120 ks   | 130 ks     | +10 ks   |

‚û°Ô∏è Sleduje se:
- absolutn√≠ chyba
- smƒõrov√° chyba (p≈ôest≈ôelen√≠ / podst≈ôelen√≠)

## 2. Nejƒçastƒõj≈°√≠ statistick√© metriky p≈ôesnosti

### üìèMAE ‚Äì Mean Absolute Error
Pr≈Ømƒõrn√° absolutn√≠ chyba

MAE ud√°v√°, o kolik jednotek se predikce v pr≈Ømƒõru li≈°√≠ od skuteƒçnosti ‚Äì bez ohledu na smƒõr chyby.

‚û°Ô∏è Nap≈ô. MAE = 8 znamen√°, ≈æe se model v pr≈Ømƒõru m√Ωl√≠ o 8 kus≈Ø.

Vzorec
$$MAE = \frac{1}{n} \sum_{i=1}^{n} \left| y_i - \hat{y}_i \right|$$

- $ y_i$ ‚Ä¶ skuteƒçn√° hodnota  
- $ \hat{y}_i$ ‚Ä¶ predikovan√° hodnota  
- $ n$ ‚Ä¶ poƒçet pozorov√°n√≠

**üêç Python k√≥d** 

````python
import numpy as np

y_true = np.array([100, 120, 130, 90])
y_pred = np.array([110, 115, 125, 95])

mae = np.mean(np.abs(y_true - y_pred))
print("MAE:", mae)
````

**üóÑÔ∏èSQL k√≥d**
```sql
SELECT 
    AVG(ABS(skutecnost - predikce)) AS MAE
FROM prodeje;

**üìä Interpretace**
| Hodnota MAE | Kvalita      |
| ----------- | ------------ |
| 0‚Äì5         | vynikaj√≠c√≠   |
| 5‚Äì10        | dobr√°        |
| 10‚Äì20       | slab≈°√≠       |
| nad 20      | nepou≈æiteln√° |


### üìèRMSE ‚Äì Root Mean Square Error

Kvadratick√° pr≈Ømƒõrn√° chyba

Je to p≈ô√≠snƒõj≈°√≠ metrika ne≈æ MAE, proto≈æe 
- umoc≈àuje chyby
- v√≠ce trest√° velk√© p≈ôest≈ôely

Pou≈æ√≠v√° se tam, kde jsou velk√© odchylky kritick√© (sklad, finance, logistika).
<br>RMSE mƒõ≈ô√≠ pr≈Ømƒõrnou velikost chyby mezi predikovan√Ωmi hodnotami $\hat{y}_i$ a skuteƒçn√Ωmi hodnotami $y_i$.

‚û°Ô∏è Pokud je RMSE v√Ωraznƒõ vƒõt≈°√≠ ne≈æ MAE, znamen√° to v√Ωskyt extr√©mn√≠ch chyb.
<br>‚û°Ô∏è Ide√°ln√≠, kdy≈æ jsou velk√© chyby kritick√© pro byznys.

Vzorec
$$ \text{RMSE} = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (\hat{y}_i - y_i)^2} $$

- $ y_i$ ‚Ä¶ skuteƒçn√° hodnota  
- $ \hat{y}_i$ ‚Ä¶ predikovan√° hodnota  
- $ n$ ‚Ä¶ poƒçet pozorov√°n√≠

**üêç Python k√≥d** 


````python
import numpy as np

y_true = np.array([100, 120, 130, 90])
y_pred = np.array([110, 115, 125, 95])

rmse = np.sqrt(np.mean((y_true - y_pred)**2))
print("RMSE:", rmse)
````


**üóÑÔ∏è SQL k√≥d** 
```sql
SELECT 
    SQRT(AVG(POWER(skutecnost - predikce, 2))) AS RMSE
FROM prodeje;

**üìä Interpretace**
| RMSE                  | V√Ωznam              |
| --------------------- | ------------------- |
| bl√≠zk√© MAE            | stabiln√≠ model      |
| v√Ωraznƒõ vy≈°≈°√≠ ne≈æ MAE | v√Ωskyt velk√Ωch chyb |
| extr√©mnƒõ vysok√©       | model selh√°v√°       |


### üìèMAPE ‚Äì Procentu√°ln√≠ chyba

Chyba v procentech

<br>‚û°Ô∏è ide√°ln√≠ pro obchodn√≠ a finanƒçn√≠ reporting
<br>‚û°Ô∏è o kolik procent se model v pr≈Ømƒõru m√Ωl√≠
<br>Nap≈ô. MAPE = 5 % = velmi dobr√° predikce

Vzorec
$$\text{MAPE} = \frac{100}{n} \sum_{i=1}^{n} \left| \frac{y_i - \hat{y}_i}{y_i} \right|$$


- $ y_i$ ‚Ä¶ skuteƒçn√° hodnota  
- $ \hat{y}_i$ ‚Ä¶ predikovan√° hodnota  
- $ n$ ‚Ä¶ poƒçet pozorov√°n√≠

**üêç Python k√≥d**

````python
import numpy as np

y_true = np.array([100, 120, 130, 90])
y_pred = np.array([110, 115, 125, 95])

mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
print("MAPE:", mape, "%")
````

**üóÑÔ∏è SQL k√≥d**
```sql
SELECT 
    AVG(ABS((skutecnost - predikce) / skutecnost)) * 100 AS MAPE
FROM prodeje
WHERE skutecnost <> 0;


**üìä Interpretace MAPE**
| MAPE     | Kvalita predikce |
| -------- | ---------------- |
| do 5 %   | excelentn√≠       |
| 5‚Äì10 %   | velmi dobr√°      |
| 10‚Äì20 %  | pou≈æiteln√°       |
| nad 20 % | slab√°            |
